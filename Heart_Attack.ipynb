# compare BMI | Heart Attack

```python
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal'
    elif 25 <= bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

```

```python
df['BMI_Category'] = df['BMI'].apply(categorize_bmi)
```

```python
bmi_heart_attack = df.groupby('BMI_Category')['Heart_Attack'].sum()
bmi_heart_attack
```

* Obese people are more likely to have heart attacks compared to others.

```python
plt.figure(figsize=(7,5))
bmi_heart_attack.plot.pie(autopct='%1.1f%%', labels=bmi_heart_attack.index)
```
